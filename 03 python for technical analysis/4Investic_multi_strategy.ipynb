```python
import pandas as pd
import yfinance as yf
import pandas_ta as ta
```

```python
start = '2021-01-01'
end = '2022-01-01'
```

```python
data = yf.download('BTC-USD', start=start, end=end)
df = data.copy()
```

```python
df.ta.macd(append=True)
df.ta.rsi(10,append=True)
```

```python
df
```

```python
CustomStrategy = ta.Strategy(
    name='anything',
    description='anyhow',
    ta = [{'kind':'ema', 'length':12},
          {'kind':'ema', 'length':26},
          {'kind':'macd', 'fast':12, 'slow':21},
          {'kind':'rsi'}]
)
```

```python
df.ta.strategy(CustomStrategy)
```

```python
df
```

```python
df['MACD_trend'] = df['MACD_12_21_9'] > df['MACDs_12_21_9']
```

```python
df.loc[(df['MACD_trend']==True) & (df['MACD_trend'].shift()==False) & df['RSI_14']>50, 'action'] = 'buy'
```

```python
df[df.action=='buy'].iloc[:,-5:]
```

```python
df.loc[(df['MACD_trend']==False)&(df['MACD_trend'].shift()==True) & (df['RSI_14'] <50), 'action'] = 'sell'
```

```python
df[df.action.notnull()].iloc[:,-5:]
```

```python
df = data.copy()
df.ta.strategy(CustomStrategy)
```
