```python
import pandas as pd
import mplfinance as mpf
import yfinance as yf
```

```python
start = '2022-08-01'
end = '2022-09-01'
```

```python
data = yf.download('BTC-USD', interval='1h',start=start, end=end)
```

```python
data.Close.plot()
```

```python
data
```

```python
df = data.iloc[:,:-2]
```

```python
df
```

## MACD Formula
## MACD= 12-Period EMA  26-Period EMA

## Signal line = 9-Period MACD EMA 

```python
df.Close.ewm(12,adjust=False,min_periods=12).mean()
```

```python
EMA_short = df['Close'].ewm(span=12, adjust=False,min_periods= 12).mean()
EMA_long = df['Close'].ewm(span=26, adjust=False,min_periods= 26).mean()
```

```python
EMA_short
```

```python
EMA_long
```

```python
MACD = EMA_short - EMA_long
MACD
```

```python
signal = MACD.ewm(span=9, adjust=False,min_periods=9).mean()
 
```

```python
signal
```

```python
histo = MACD - signal
```

```python
histo
```

```python
df['macd'] = MACD
df['signal_line'] = signal
df['histogram'] = df['macd'] - df['signal_line']
```

```python
df
```

image.png

```python
df['macd'] > df['signal_line']
```

```python
df['trend']  = df['histogram'] > 0
```

```python
df
```

```python
df['shift_hist'] = df['histogram'].shift(1)
```

```python
df[(df['histogram'] < 0) & (df['shift_hist'] > 0)]
```

```python
#trend 
df['trend'] = df['macd'] > df['signal_line']
```

```python
df.loc[(df['trend'] == True) & (df['trend'].shift()==False),'action'] = 'buy'
df.loc[(df['trend'] == False) & (df['trend'].shift()==True),'action'] = 'sell'
#same result block on bottom
```

```python
df.loc[(df['histogram'] > 0) & (df['shift_hist'] < 0),'action'] = 'buy'
df.loc[(df['histogram'] < 0) & (df['shift_hist'] > 0),'action'] = 'sell'
```

```python
df
```

```python
df[df.action.notnull()]
```

# Visualization

```python
#for plot MACD
mpf_macd = mpf.make_addplot(df['macd'],panel = 1,color = 'blue', title ='MACD')
mpf_signal = mpf.make_addplot(df['signal_line'],panel = 1,color = 'orange')

colors = ['green' if x >= 0 else 'red' for x in df["histogram"]]
mpf_histogram = mpf.make_addplot(df['histogram'],type='bar',panel = 1,color = colors)

macd_plot = [mpf_macd,mpf_signal,mpf_histogram]
```

```python
type(mpf_histogram)
```

```python
type(macd_plot)
```

```python
mpf.plot(df,type='candle',style='yahoo',addplot=macd_plot)
```

```python
#  marker
df.loc[df['action']=='buy','marker_position'] = df['Low']*0.95
df.loc[df['action']=='sell','marker_position'] = df['High']*1.05
```

```python
#for buy, sell signal
markers = ['^' if x == 'buy' else 'v' for x in df['action']] 
color_marker = ['green' if x == 'buy' else 'red' for x in df['action']]
plot_signal = mpf.make_addplot(df['marker_position'],type='scatter',markersize= 30, marker=markers,color=color_marker)
```

```python
type(plot_signal)
```

```python
all_plot = macd_plot + [plot_signal]
```

```python
mpf.plot(df,type='candle',style='yahoo',addplot=all_plot)
```
