```python
import pandas_ta as ta
import yfinance as yf
import mplfinance as mpf
```

```python
data = yf.download('BTC-USD', interval='1d',start='2021-01-01', end='2022-01-01')
```

```python
data.ta.macd(fast=12,slow=21,signal=7)
```

```python
df_ta = data.copy()
```

```python
df_ta.ta.macd(slow=21,append=True)
```

```python
df_ta
```

```python
rsi = df_ta.ta.rsi()
```

```python
df_ta.ta.rsi(append=True)
df_ta
```

```python
df_ta.columns = ['Open',	'High',	'Low',	'Close',	'Adj Close',	'Volume',	'macd',	'histogram',	'signal_line',	'rsi']
```

```python
#for plot MACD
mpf_macd = mpf.make_addplot(df_ta['macd'],panel = 1,color = 'blue',ylabel='macd')
mpf_signal = mpf.make_addplot(df_ta['signal_line'],panel = 1,color = 'orange')

colors = ['green' if x >= 0 else 'red' for x in df_ta["histogram"]]
mpf_histogram = mpf.make_addplot(df_ta['histogram'],type='bar',panel = 1,color = colors)

mpf_rsi = mpf.make_addplot(df_ta['rsi'],panel=2,color='orange',ylabel='rsi')

ind_plot = [mpf_macd,mpf_signal,mpf_histogram,mpf_rsi]
```

```python
mpf.plot(df_ta,type='candle',style='default',addplot=ind_plot,panel_ratios=(3,1,1),figratio=(2,1),figscale=1.5)
```

```python
mpf.plot(df_ta,type='candle',style='default',addplot=ind_plot, hlines = dict(hlines=[65000, 30000], colors=['r','g'], linestyle='-'))
```

```python
df_ta['upper'] = 70
df_ta['lower'] = 30
```
